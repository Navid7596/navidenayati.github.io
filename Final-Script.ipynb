```python
import geopandas as gpd
import matplotlib.pyplot as plt
```

```python
# Load the Blighted_Area and Dumping2024 shapefiles from the new folder
blighted_areas = gpd.read_file(r"C:\Users\simani\OneDrive - The University of Memphis\Attachments\Navid\OUTPUT\blighted_areas.shp")
litter_data = gpd.read_file(r"C:\Users\simani\OneDrive - The University of Memphis\Attachments\Navid\OUTPUT\litter_data.shp")
```

```python
# Reproject to a projected CRS (e.g., EPSG:3857 for Web Mercator)
blighted_areas_projected = blighted_areas.to_crs(epsg=3857)
litter_data_projected = litter_data.to_crs(epsg=3857)
```

```python
# Create a buffer around each blighted area (e.g., 100 meters)
blighted_areas_projected['buffer'] = blighted_areas_projected.geometry.buffer(100)
```

```python
# Plot the original blighted areas and their buffers
fig, ax = plt.subplots(figsize=(10, 10))
blighted_areas_projected.plot(ax=ax, color='blue', alpha=0.5, edgecolor='black', label="Blighted Areas")
blighted_areas_projected['buffer'].plot(ax=ax, color='red', alpha=0.3, label="Buffer (100m)")
plt.title("Blighted Areas with 100m Buffer")
plt.legend()
plt.show()
```

```python
# Define chunk size
chunk_size = 500  # Adjust based on available memory

# Initialize empty list to store join results
join_results = []

# Process blighted areas in chunks
for i in range(0, len(blighted_areas_projected), chunk_size):
    chunk = blighted_areas_projected.iloc[i:i + chunk_size]
    
    # Perform the spatial join for each chunk
    for _, blight_row in chunk.iterrows():
        # Find litter points within the buffer of the blighted area
        within_buffer = litter_data_projected[litter_data_projected.geometry.within(blight_row['buffer'])]
        
        for _, litter_row in within_buffer.iterrows():
            join_results.append({
                'litter_id': litter_row.name,
                'blighted_area_id': blight_row.name,
                'geometry': litter_row['geometry']
            })

# Convert the list of join results into a GeoDataFrame
join_result = gpd.GeoDataFrame(join_results)

# Display the first few rows of the joined data
print(join_result.head())
```

```python
# Calculate the distance between litter points and blighted area buffers
join_result['distance_to_blighted'] = join_result['geometry'].distance(blighted_areas_projected['buffer'].iloc[0])
```

```python
# Perform Pearson correlation between litter count and proximity to blighted areas
correlation = join_result['distance_to_blighted'].corr(join_result['litter_id'])  # Example of correlation, adjust as needed
print(f"Pearson correlation: {correlation}")
```

```python
# Scatter plot of litter count vs. distance to blighted areas
plt.scatter(join_result['distance_to_blighted'], join_result['litter_id'], alpha=0.5)
plt.xlabel('Distance to Blighted Areas (m)')
plt.ylabel('Litter Count')
plt.title('Litter Count vs. Distance to Blighted Areas')
plt.show()
```

```python
# Plot the join result: Litter points within the blighted area buffer
ax = join_result.plot(marker='o', color='red', markersize=5, label="Litter Points")
blighted_areas_projected.plot(ax=ax, color='blue', alpha=0.5, edgecolor='black', label="Blighted Areas")
plt.title("Spatial Join: Litter Points within Blighted Areas")
plt.legend()
plt.show()
```
